B

```python
covar_matrix=np.cov(sc_data.T)
```

```python
eigen_values,eigen_vector=np.linalg.eig(covar_matrix)
```

```python
summ=sum(eigen_values)
variance_explained=[(i/summ)*100 for i in sorted(eigen_values,reverse=True)]
cum_sum=np.cumsum(variance_explained)
```

```python
print('The variance explained by each components are : \n',cum_sum)
```

```python
print('The number of components to be used so that we can get 90% variance explained is ',len(cum_sum[cum_sum<=90]))
```

```python
plt.figure(figsize=(8,6))
plt.bar(range(31),variance_explained,alpha=0.5,align='center',label='Variance explained by each')
plt.step(range(31),cum_sum,alpha=0.5,where='mid',label='cumulative variance')
plt.ylabel('ratio of explained variance')
plt.xlabel('Principal components')
plt.legend(loc='upper left')
plt.tight_layout
plt.show()
```

```python
from sklearn.decomposition import PCA
```

* as we can see in the above boxplot there are ouliers 

* proceeding with capping because kmeans is affected by outliers 
and also we get better silloute score if we do oulier treatment

```python
for i in pca_data.columns:
    q1=pca_data[i].quantile(0.25)
    q3=pca_data[i].quantile(0.75)
    iqr=q3-q1
    ub=q3+(1.5*iqr)
    lb=q1-(1.5*iqr)
    uc=pca_data[i].quantile(0.99)     #treating outliers
    lc=pca_data[i].quantile(0.01)
    for j in pca_data.index:
        if pca_data.loc[j,i]>ub:
            pca_data.loc[j,i]=uc
        if pca_data.loc[j,i]<lb:
            pca_data.loc[j,i]=lc
```

```python
for i in range(0,14):
    sns.boxplot(pca_data[i])
    plt.title('PCA_'+str(i))
    plt.show()
```

#### finding the optimun kvalue with the help of elbow curve

randomly applying kmeans clustering with range of 1-10 as k values and 

saving errors by that model for that particular kvalue

```python
from sklearn.cluster import KMeans
wcsse=[]
for i in range(1,10):
    km=KMeans(n_clusters=i,random_state=0)
    km.fit(pca_data)
    wcsse.append(km.inertia_)
```

plotting the same 

```python
clusters=pd.DataFrame({'cluster':range(1,10),'wcsse':wcsse})
plt.plot(clusters['cluster'],clusters['wcsse'],'-o')
```

from the above we can see that when cluser is 2 the curve is dropping 

hence when n clusters are 2 the drop is more in errors

and also since elbow curve will only se the distance within the clusters 

we can see the siloutee scores the k values 

#### finding the optimun kvalue with the help of sillhoutte scores

```python
from sklearn.metrics import silhouette_score
for i in range(2,10):
    km=KMeans(n_clusters=i,random_state=0)
    km.fit(pca_data)
    labels=km.predict(pca_data)
    print('The sillhoutte score for ',i,' clusters :',silhouette_score(pca_data,labels))
```

from the above  it is clear that when n_clusters are 2 the sillhoute score is high 

#### visualizing the number of cluster approx'ly with help of dendrogram

```python
from scipy.cluster.hierarchy import linkage,dendrogram
ward_link=linkage(pca_data,method='ward') # ward is the best method 
dendrogram(ward_link,leaf_rotation=90.0)
plt.tight_layout()
plt.show()
```

from the above visualization it is clear that there are 2 clusters formed
